```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
f = 'transactions_train.csv'
df = pd.read_csv(f)
```

```python
from datetime import datetime
import calendar

def to_date(dt):
    dat = datetime.strptime(dt, '%Y-%m-%d')
#     print(dat)
    return calendar.day_name[dat.weekday()]

def to_season(dt):
    dat = datetime.strptime(dt, '%Y-%m-%d')
    month = dat.month
    if(month < 3):
        return 'Winter'
    elif(month < 6):
        return 'Spring'
    elif(month < 9):
        return 'Summer'
    elif(month < 12):
        return 'Fall'
    else:
        return 'Winter'
    print(dat.month)
to_date('2018-09-20')
to_season('2018-09-20')

# my_date = date.today()
# calendar.day_name[my_date.weekday()]  #'Wednesday'
```

```python
actualpd = df
```

```python
actualpd['weekday'] = actualpd['t_dat'].apply(to_date)
actualpd['Season'] = actualpd['t_dat'].apply(to_season)
most_common = actualpd.groupby(['customer_id']).agg(lambda x:x.value_counts().index[0])
```

```python
most_common
```

```python
time_most_common = most_common[['weekday', 'Season']]
time_most_common
```
